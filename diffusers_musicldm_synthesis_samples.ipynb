```python
from diffusers import AudioLDMPipeline
from IPython.display import Audio
import torch

repo_id = "reach-vb/mldm-test"
pipe = AudioLDMPipeline.from_pretrained(repo_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompt = "Western music, chill out, folk instrument R & B beat"
audio = pipe(prompt, num_inference_steps=50, audio_length_in_s=10.0).audios[0]
Audio(audio, rate=16000)
```
