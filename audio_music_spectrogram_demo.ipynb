```python
from diffusers import SpectrogramDiffusionPipeline, MidiProcessor

pipe = SpectrogramDiffusionPipeline.from_pretrained("google/music-spectrogram-diffusion")
pipe = pipe.to("cuda")
processor = MidiProcessor()
```

```python
!wget http://www.piano-midi.de/midis/beethoven/beethoven_hammerklavier_2.mid
```

```python
output = pipe(processor("beethoven_hammerklavier_2.mid")[:2])

audio = output.audios[:2]
```
