```python
from huggingface_hub import InferenceClient
```

```python
client = InferenceClient(model = "https://zj5lt7pmzqzbp0d4.us-east-1.aws.endpoints.huggingface.cloud")
```

```python
response = client.post(json={"inputs": "happy folk song, cheerful and lively"})
```

```python
from IPython.display import Audio

Audio(eval(response)[0]["generated_audio"], rate=32000)
```

```python
output = eval(response)[0]["generated_audio"]
```

```python
import scipy
import numpy as np

scipy.io.wavfile.write("musicgen_out.wav", rate=32000, data=np.array(output[0]))
```
